```python
import pickle
import pprint
import math

def get_driver_stats(driver):
    #get all driver names
    #if driver in driver names
    #then filter by its name in dataframe
    if driver in(df_drivers['last_name'].unique()):
        df_drivers_filtered = df_drivers[df_drivers['last_name'] == driver].copy()
        df_races_filtered = df_race_results[df_race_results['last_name'] == driver].copy()
        df_sprints_filtered = df_sprint_results[df_sprint_results['last_name'] == driver].copy()

        df_drivers_filtered['team'] = df_races_filtered['team'].unique()
        df_drivers_filtered['total_points_races'] = df_races_filtered['points'].sum()
        df_drivers_filtered['total_races_completed'] = df_races_filtered[df_races_filtered['status'] == 'Finished']['circuit'].count()
        df_drivers_filtered['total_points_sprint'] = df_sprints_filtered['points'].sum()
        df_drivers_filtered['total_pole_positions'] = df_races_filtered[df_races_filtered['starting_position'] == 1]['circuit'].count()
        df_drivers_filtered['total_dnfs'] = df_races_filtered[~df_races_filtered['status'].isin(['Finished', '+1 Lap', '+2 Laps'])]['circuit'].count()
        df_drivers_filtered['total_crashes'] = df_races_filtered[df_races_filtered['status'].isin(['Collision', 'Collision damage' 'Accident'])]['circuit'].count()
        df_drivers_filtered['dnf_ratio'] = round((df_drivers_filtered['total_dnfs']/df_races_filtered['circuit'].count())* 100, 2).astype(str) + '%'
        df_drivers_filtered['crash_ratio'] = round((df_drivers_filtered['total_crashes']/df_races_filtered['circuit'].count())* 100, 2).astype(str) + '%'
        df_drivers_filtered['avg_points_per_race'] = math.trunc(np.ceil(df_races_filtered['points'].mean()))
        df_drivers_filtered['avg_points_per_sprint'] = math.trunc(np.ceil(df_sprints_filtered['points'].mean()))
        df_drivers_filtered['avg_qualifying_position'] = math.trunc(np.ceil(df_races_filtered['starting_position'].mean()))
        df_drivers_filtered['avg_ending_position_race'] = math.trunc(np.ceil(df_races_filtered['ending_position'].mean()))
        df_drivers_filtered['avg_ending_position_sprint'] = math.trunc(df_sprints_filtered['position'].mean())
        df_drivers_filtered['avg_points_per_sprint'] = math.trunc(np.ceil(df_sprints_filtered['points'].mean()))
  
        return df_drivers_filtered
    else:
        print("Please type a correct driver name")
```

```python
import pickle
import pprint
import math

def get_all_driver_stats(df_drivers):
    
    all_driver_dfs = []

    # get all driver surnames
    # loop through them and run the code below
    # append dataframe

    for driver in(df_drivers['last_name'].unique()):
        df_drivers_filtered = df_drivers[df_drivers['last_name'] == driver].copy()
        df_races_filtered = df_race_results[df_race_results['last_name'] == driver].copy()
        df_sprints_filtered = df_sprint_results[df_sprint_results['last_name'] == driver].copy()
        
        df_drivers_filtered['team'] = df_races_filtered['team'].unique()
        df_drivers_filtered['total_points_races'] = df_races_filtered['points'].sum()
        df_drivers_filtered['total_races_completed'] = df_races_filtered[df_races_filtered['status'] == 'Finished']['circuit'].count()
        df_drivers_filtered['total_points_sprint'] = df_sprints_filtered['points'].sum()
        df_drivers_filtered['total_pole_positions'] = df_races_filtered[df_races_filtered['starting_position'] == 1]['circuit'].count()
        df_drivers_filtered['total_dnfs'] = df_races_filtered[~df_races_filtered['status'].isin(['Finished', '+1 Lap', '+2 Laps'])]['circuit'].count()
        df_drivers_filtered['total_crashes'] = df_races_filtered[df_races_filtered['status'].isin(['Collision', 'Collision damage' 'Accident'])]['circuit'].count()
        df_drivers_filtered['dnf_ratio'] = round((df_drivers_filtered['total_dnfs']/df_races_filtered['circuit'].count())* 100, 2).astype(str) + '%'
        df_drivers_filtered['crash_ratio'] = round((df_drivers_filtered['total_crashes']/df_races_filtered['circuit'].count())* 100, 2).astype(str) + '%'
        df_drivers_filtered['avg_points_per_race'] = math.trunc(np.ceil(df_races_filtered['points'].mean()))
        df_drivers_filtered['avg_points_per_sprint'] = math.trunc(np.ceil(df_sprints_filtered['points'].mean()))
        df_drivers_filtered['avg_qualifying_position'] = math.trunc(np.ceil(df_races_filtered['starting_position'].mean()))
        df_drivers_filtered['avg_ending_position_race'] = math.trunc(np.ceil(df_races_filtered['ending_position'].mean()))
        df_drivers_filtered['avg_ending_position_sprint'] = math.trunc(df_sprints_filtered['position'].mean())
        df_drivers_filtered['avg_points_per_sprint'] = math.trunc(np.ceil(df_sprints_filtered['points'].mean()))

        all_driver_dfs.append(df_drivers_filtered)

    all_drivers_merged_df = pd.concat(all_driver_dfs, ignore_index=True)        

    return all_drivers_merged_df
```

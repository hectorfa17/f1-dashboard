```python
race_results_list = []

# Loop through each race in the race results
try:
    for x in range(len(race_results['MRData']['RaceTable']['Races'])):
        races = race_results['MRData']['RaceTable']['Races'][x]
        season = races.get('season', "")
        round = races.get('round', "")
        circuit =  races.get('raceName', "")
        date = races.get('date', "")
        # time = races.get('time', "")
        
        for y in range(len(race_results['MRData']['RaceTable']['Races'][x]['Results'])):
            results = race_results['MRData']['RaceTable']['Races'][x]['Results'][y]
            laps = results.get('laps', '')
            driver = results.get('Driver', {})
            first_name = driver.get('givenName', "")
            last_name = driver.get('familyName', "")
            driver_code = driver.get('code', "")
            driver_date_of_birth = driver.get('dateOfBirth', "")
            driver_nationality = driver.get('nationality', "")
            driver_permanent_number = driver.get('permanentNumber', "")
            driver_info = driver.get('url', "")
            starting_position =  results.get('grid', "")
            ending_position = results.get('position', "")
            points = results.get('points', "")
            status = results.get('status', "")
            points = results.get('points', "")
            time_stats = results.get('Time', {})
            finish_time = time_stats.get('time', "")
            finish_time_ms = time_stats.get('millis', "")
            fastest_lap_stats = results.get('FastestLap', {'rank': "", 'lap': "", 'Time': {'time': ""}, 'AverageSpeed': {'speed': ""}})
            fastest_lap_time = fastest_lap_stats['Time'].get('time', "")
            fastest_lap_avg_speed = fastest_lap_stats['AverageSpeed'].get('speed', "")
            fastest_lap_rank = fastest_lap_stats.get('rank', "")
            fastest_lap_lap = fastest_lap_stats.get('lap', "")
            constructor =  results.get('Constructor', {})
            team = constructor.get('name', "")
            team_id = constructor.get('constructorId', "")
            # team_nationality = constructor.get('nationality', "")
            # team_info = constructor.get('url', "")

            race_results_dict = {   
                                    'season': season,
                                    'round': round,
                                    'circuit': circuit,
                                    'date': date,
                                    # 'time': time,        
                                    'laps': laps,
                                    'first_name': first_name,
                                    'last_name': last_name,
                                    'driver_code': driver_code,
                                    #'driver_date_of_birth': driver_date_of_birth,
                                    #'driver_nationality': driver_nationality,
                                    #'driver_perm_number': driver_permanent_number,
                                    #'driver_info': driver_info,
                                    'starting_position': starting_position,
                                    'ending_position': ending_position,
                                    'points': points,
                                    'status': status,
                                    'finish_time': finish_time,
                                    'finish_time_ms': finish_time_ms,
                                    'fastest_lap_time': fastest_lap_time,
                                    'fastest_lap_avg_speed': fastest_lap_avg_speed,
                                    'fastest_lap_rank': fastest_lap_rank,
                                    'fastest_lap_lap': fastest_lap_lap,
                                    'team': team,
                                    'team_id': team_id,
                                    #'team_nationality': team_nationality,
                                    #'team_info': team_info
            }
            
            race_results_list.append(race_results_dict.copy())

except IndexError:
    print('no more races to fetch data from')
```
